```python
import pandas as pd
import sklearn
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python

dataset = pd.read_csv("/home/aswnss/Coding/Python/Machine-Learning-Projects/ras ml/datasets/diabetes.csv",sep=',')
dataset
```

```python
dataset.groupby('Outcome').mean()
```

```python
dataset.isna().sum()
```

```python
# dropping Pregnancies SkinThickness DiabetesPedigreeFunction
dataset2 = dataset.drop(['Pregnancies','BloodPressure','DiabetesPedigreeFunction'],axis=1)
dataset2
```

```python
X = dataset2.drop(['Outcome'],axis=1)
y = dataset2.Outcome
D_tree = sklearn.tree.DecisionTreeClassifier()
```

```python
max_acc = -1
for _ in range(500):
    x_train,x_test,y_train,y_test= sklearn.model_selection.train_test_split(X,y,test_size=0.1)
    D_tree.fit(x_train,y_train)
    acc = D_tree.score(x_test,y_test)
    if acc > max_acc:
        max_acc = acc
print(max_acc*100)
```

```python
R_forest = RandomForestClassifier(n_estimators=100)
```

```python
max_acc = -1
for _ in range(100):
    x_train,x_test,y_train,y_test= sklearn.model_selection.train_test_split(X,y,test_size=0.1)
    R_forest.fit(x_train,y_train)
    acc = R_forest.score(x_test,y_test)
    if acc > max_acc:
        max_acc = acc
print(max_acc*100)
```
